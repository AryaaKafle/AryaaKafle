```
# 2.15
x = float(input("Enter first float: "))
y = float(input("Enter second float: "))
z = float(input("Enter third float: "))

# Determine ascending order using comparisons (works even with duplicates).
if x <= y:
    if y <= z:
        a, b, c = x, y, z
    elif x <= z:
        a, b, c = x, z, y
    else:
        a, b, c = z, x, y
else:
    if x <= z:
        a, b, c = y, x, z
    elif y <= z:
        a, b, c = y, z, x
    else:
        a, b, c = z, y, x

print("Ascending order:", a, b, c)
```

```
# 3.1
passes = 0
total_students = 10
student = 1

while student <= total_students:
    result = input("Enter result (1=pass, 2=fail): ").strip()
    if result not in ("1", "2"):
        print("Invalid input. Please enter 1 or 2.")
        continue  # ask again for the same student

    if result == "1":
        passes += 1

    student += 1

failures = total_students - passes

print("Passed:", passes)
print("Failed:", failures)

if passes > 8:
    print("Bonus to instructor!")
```

```
# 3.11
total_miles = 0.0
total_gallons = 0.0

while True:
    miles = float(input("Enter miles driven (-1 to end): "))
    if miles == -1:
        break

    gallons = float(input("Enter gallons used: "))
    if gallons <= 0:
        print("Gallons must be > 0. Try again.")
        continue

    mpg = miles / gallons
    print(f"MPG this tankful: {mpg:.2f}")

    total_miles += miles
    total_gallons += gallons

if total_gallons > 0:
    combined_mpg = total_miles / total_gallons
    print(f"\nTotal miles: {total_miles:.2f}")
    print(f"Total gallons: {total_gallons:.2f}")
    print(f"Combined MPG: {combined_mpg:.2f}")
else:
    print("No data entered.")
```

```
# 3.17
rows = 10

# Pattern (a): 1 to 10 asterisks
for i in range(1, rows + 1):
    print('*' * i)
print()

# Pattern (b): 10 down to 1 asterisks
for i in range(rows, 0, -1):
    print('*' * i)
print()

# Pattern (c): right-aligned increasing
for i in range(1, rows + 1):
    spaces = rows - i
    print(' ' * spaces + '*' * i)
print()

# Pattern (d): right-aligned decreasing
for i in range(rows, 0, -1):
    spaces = rows - i
    print(' ' * spaces + '*' * i)
```

```
# 3.18
rows = 10
sep = "   "

for i in range(1, rows + 1):
    # (a) increasing
    a = '*' * i

    # (b) decreasing
    b = '*' * (rows - i + 1)

    # (c) right-aligned increasing
    c = ' ' * (rows - i) + '*' * i

    # (d) right-aligned decreasing
    d = ' ' * (i - 1) + '*' * (rows - i + 1)

    print(f"{a:<10}{sep}{b:<10}{sep}{c:<10}{sep}{d:<10}")
```

```
# 3.19
limit = 20
triples = []

for side1 in range(1, limit + 1):
    for side2 in range(side1, limit + 1):  # side2 >= side1 to reduce duplicates
        for hypotenuse in range(side2, limit + 1):  # hypotenuse >= side2
            if side1**2 + side2**2 == hypotenuse**2:
                triples.append((side1, side2, hypotenuse))

print("Pythagorean triples with values ≤ 20:")
for t in triples:
    print(t)
```

```
# 3.20
binary = int(input("Enter a binary number (e.g., 1101): "))

decimal = 0
position_value = 1

while binary > 0:
    digit = binary % 10
    if digit not in (0, 1):
        raise ValueError("Input contained a digit other than 0 or 1.")
    decimal += digit * position_value
    position_value *= 2
    binary //= 10

print("Decimal equivalent:", decimal)
```
